Email Validator (String Filtering + Loops)

üíº Real-world use: Used in signup forms or data cleaning tasks.

Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.

Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
#email Validator
email = input("Enter your email: ")
#check: contains "@"
has_at = False
for character in email:
    if character == "@":
        has_at = True
        break
#check ending
end_character = False

#last 4 character for .com
if email[-4:] == ".com":
    end_character = True

#last 4 character for .org
elif email[-4:] == ".org":
    end_character = True

if has_at and end_character:
    print("Valid Email")

else:
    print("Invalid Email")



```

Problem:02

üíº Real-world use: Analyzing reviews or comments in text analytics.

Task:
Given a review text, count how many times each word appears (ignore case).

Example:
Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
text = input("Enter review: ")

# Convert to lowercase to ignore case
text = text.lower()

# Split text into words
words = text.split()

# Empty dictionary to store counts
frequency = {}

# Count each word using loop
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)
```

problem:03

 Password Strength Checker
üíº Real-world use: Used in user authentication systems.

Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:

Input: "Hello@123"
Output: Strong Password

```python
password = input("Enter your password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

for ch in password:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    else:
        has_special = True

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

problrm 4: Username Generator

üíº Real-world use: Automatically create usernames for users.

Task:

Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

Example:

Input: "Mamun Malitha"
Output: mamitha57

```python
import random

full_name = input("Enter your full name: ")

# Split name
parts = full_name.split()
first = parts[0]
last = parts[-1]

# Take first 3 letters of first name using loop
first_part = ""
i = 0
while i < 3 and i < len(first):
    first_part += first[i]
    i += 1

# Take last 3 letters of last name using loop
last_part = ""
j = len(last) - 3
if j < 0:
    j = 0

while j < len(last):
    last_part += last[j]
    j += 1

# Generate random number (10‚Äì99)
num = random.randint(10, 99)

# Final username
username = first_part + last_part + str(num)

print("Generated Username:", username.lower())
```

Problrm o5: SMS Word Counter

üíº Real-world use: Used in messaging apps or social media post limit check.

Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit

```python
message = input("Enter your message: ")

# Count characters using loop
char_count = 0
for ch in message:
    char_count += 1

# Count words using loop
word_count = 1   # words = spaces + 1
for ch in message:
    if ch == " ":
        word_count += 1

# Check SMS limit (160 characters)
if char_count > 160:
    status = "Exceeds Limit"
else:
    status = "Within Limit"

print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
```

Problrm 06:  Remove Duplicates from Contact List

üíº Real-world use: Data cleaning in CRM or user databases.

Task:

Given a list of names (comma separated), remove duplicates.

Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
contacts = input("Enter names (comma separated): ")

# Split names into a list
names = contacts.split(",")

unique_list = []   # store unique names

# Loop through names
for name in names:
    clean_name = name.strip()   # remove extra spaces
    
    # add only if not already in unique_list
    if clean_name not in unique_list:
        unique_list.append(clean_name)

# Join result back into string
result = ", ".join(unique_list)

print("Unique Contacts:", result)
```

Problrm 07: Detect Suspicious Words in Text (Moderation)

üíº Real-world use: Used in chat moderation or email filtering.

Task:

If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:

Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
text = input("Enter your text: ")

banned_words = ["spam", "scam", "fake"]
found = False   # to track if any banned word is found

# Convert text to lowercase so checking becomes easy
text = text.lower()

# Loop through each banned word
for word in banned_words:
    if word in text:      # check if banned word exists
        found = True

if found:
    print("Warning: Suspicious content detected")
else:
    print("No suspicious content found")
```

Problem 08: Invoice Code Formatter
üíº Real-world use: Formatting IDs or codes in finance or ERP systems.

Task:

Given an invoice number like 23, generate code in the format INV00023.

Example:

Input: 23
Output: INV00023

```python
number = int(input("Enter invoice number: "))

# Convert number to string
num_str = str(number)

# Add leading zeros using loop
zeros_needed = 5 - len(num_str)
prefix = ""

i = 0
while i < zeros_needed:
    prefix += "0"
    i += 1

invoice_code = "INV" + prefix + num_str

print("Invoice Code:", invoice_code)
```

Problem 09: Count Hashtags and Mentions (Social Media Analytics)

üíº Real-world use: Used in analyzing social media posts.

Task:
Count how many hashtags (#) and mentions (@) appear in a post.

Example:

Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1

```python
post = input("Enter your social media post: ")

hashtag_count = 0
mention_count = 0

# Loop through each character
for ch in post:
    if ch == "#":
        hashtag_count += 1
    elif ch == "@":
        mention_count += 1

print("Hashtags =", hashtag_count)
print("Mentions =", mention_count)
```

Problem 10: Word Reversal for Data Encryption

üíº Real-world use: Simple text obfuscation or playful encoding.

Task:

Reverse each word in a sentence while keeping word order the same.

Example:

Input: "AI is fun"
Output: "IA si nuf"

```python
sentence = input("Enter a sentence: ")

words = sentence.split()   # split into words
result = []                # store reversed words

# Loop through each word
for word in words:
    reversed_word = ""
    
    # Reverse each word using loop
    i = len(word) - 1
    while i >= 0:
        reversed_word += word[i]
        i -= 1
    
    result.append(reversed_word)

# Join back into a sentence
final_sentence = " ".join(result)

print(final_sentence)
```
